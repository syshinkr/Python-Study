```python
import sqlite3

# DB 파일이 저장될 경로와 파일 이름을 써서 DB에 연결
conn = sqlite3.connect("examples/db.sqlite")

# conn.text_factory = str

#메모리에서 직접 DB 이용 코드
# conn = sqlite3.connect(":memory:")

# DB 동작을 위한 Cursor 객체 생성
# DB 사용을 위한 마지막 준비
cur = conn.cursor()
```

```python
cur.execute("""drop table if exists hanbit_books""")

cur.execute("""create table hanbit_books (
                title varchar(100),
                author varchar(100),
                translator varchar(100),
                pub_date date,
                isbn varchar(100)
                )""")
conn.commit()
```

```python
cur.execute("insert into hanbit_books values(?, ?, ?, ?, ?)",
           ("책 이름", "저자 이름", "번역자 이름", "2016-06-22", "9788968480011"))
```

### 변수 하나가 여러 열에 들어갈 때 이름 있는 변수 사용하기

```python
query_str = "insert into hanbit_books values(:title, :title, :title, :pub_date, :isbn)"
params = {
    "title":"책 이름",
    "pub_date":"2017-10-12",
    "isbn":9788968480022
}

cur.execute(query_str, params)
```

```python
import csv

csv_file = open("hanbit_media/book_list.csv", 'rt', encoding='UTF8')
csv_reader = csv.reader(csv_file)
book_list = list(csv_reader)

# csv header 제거
book_list = book_list[1:]

# 최초 크롤링 시 있었던 저자 혹은 번역자 데이터의 앞뒤 공백 제거
for item in book_list:
    item[1] = item[1].strip()
    item[2] = item[2].strip()
    
book_list
```

```python
cur.executemany("insert into hanbit_books values (?, ?, ?, ?, ?)", book_list)
```

```python
conn.commit()
```

```python
cur.execute("select * from hanbit_books where author = ?", ("이상교",))
print(cur.fetchone())
```

### 튜플을 이용한 데이터 선택

```python
query_str = "select * from hanbit_books where author=:name"
params = {
    "name":"이상교"
}

cur.execute(query_str, params)
```

#### 지정한 숫자만큼 가져오기  : fetchmany(size={숫자}) , 전부 가져오기 fetchall()

```python
result = list(cur.fetchall())

for row in result:
    print(row)
```

```python
query_str = "update hanbit_books set isbn=:isbn where author=:name"
params = {
    "isbn":"9788968480033",
    "name":"이상교"
}

cur.execute(query_str, params)
```
