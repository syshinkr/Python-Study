```python
t1 = pd.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "left_val": ["a", "b", "c", "d", "e"]
})

t1
```

```python
t2 = pd.DataFrame({
    "id": [3, 4, 5, 6, 7],
    "right_val": ["q", "w", "e", "r", "t"]
})

t2
```

```python
# merge 는 기본적으로 양쪽에 다 데이터가 있는 행만을 대상으로 함
pd.merge(t1, t2)
```

```python
# id 열을 명시적 키로 지정
pd.merge(t1, t2, on='id')
```

```python
# 양쪽 DataFrame 타입에서 서로 다른 열을 기준으로 삼기
pd.merge(t1, t2, left_on="left_val", right_on="right_val")
```

```python
# how : 어느 한쪽에 데이터가 비어도 병합할 때
# 왼쪽기준 (t1)
pd.merge(t1, t2, how="left")
```

```python
# 오른쪽 기준 (t2)
pd.merge(t1, t2, how="right")
```

```python
# 두 dataframe 타입의 모든 데이터를 순서에 맞게 병합
pd.merge(t1, t2, how="outer")
```

```python
by_health = dummy_data.groupby("health")
by_health
```

```python
# count() : 특정 열로 데이터를 묶었을 때 행 갯수
by_health.count().head()
```

```python
# size() : 특정 열로 그룹화한 행 갯수
by_health.size().head()
```

```python
# mean() : 그룹 각 열 평균값
# median() : 그룹 각 열 중간값
# sum() : 합
# min() : 최솟값
# max() : 최댓값
# 등등

by_health.max().head()
```

```python
# 여러 개의 인덱스로 그룹화
# position 열로 먼저 그룹화하고, 그 안에서 health 열을 한 번 더 그룹화
multi_indexed = heroes_df.groupby(['position', 'health'])

multi_indexed.describe()
```

```python
# get_group() : 특정 그룹의 데이터들만 DataFrame 타입 데이터로 가져오기
by_health.get_group(200)
```

```python
# agg() : 특정 열에 대해 여러가지 연산
# 파이썬 딕셔너리 타입이 파라미터
by_health.agg(
    {
        # 적용할 열 이름을 딕셔너리의 키로 넣음
        "id":
            [
                # ("<레이블이름>", <적용할 함수>)
                ("합계", np.sum),
                # <적용할 함수>
                np.mean
            ]
    }
)
```

```python
# 레이블 이름, 함수를 튜플로 묶어 전달하거나 함수 이름만 전달하기 또한 가능
by_health.agg(
    {
        "id":
            [
                np.size, np.median
            ]
    }
)
```

```python
# 제곱합 구하기
def sum_of_pow(s:pd.core.series.Series):
    # 각 그룹의 해당 열 Series 타입 데이터로 전달 받음
    return sum([1 ** 2 for i in s])


# 튜플로 열 이름을 지정하지 않고 그냥 함수로만 전달하면, 함수 이름을 열 이름으로 사용
by_health.agg({"id": [sum_of_pow]}).tail()
```

```python
# apply() : agg() 로 하기 벅찬 여러 개의 열에 작업
def get_diff_id(df : pd.core.frame.DataFrame):
    # apply() 로 전달되는 함수의 파라미터는 각 그룹의 DataFrame 타입 데이터
    return df.id.max() - df.id.min()


by_health.apply(get_diff_id)
```
